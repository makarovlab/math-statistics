```python
import pandas as pd
import seaborn as sns
```

#### Load dataset

```python
df = pd.read_csv("datasets/winequality-red.csv")
df.head()
```

#### Dataset Description

##### Context

This datasets is related to red variants of the Portuguese "Vinho Verde" wine. For more details, consult the reference [Cortez et al., 2009]. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

##### Columns

|column name|description|translation|
|-|-|-|
|fixed acidity|most acids involved with wine or fixed or nonvolatile (do not evaporate readily)|большинство кислот, связанных с вином, либо фиксированные, либо нелетучие (легко не испаряются)|
|volatile acidity|the amount of acetic acid in wine, which at too high of levels can lead to an unpleasant, vinegar taste|количество уксусной кислоты в вине, которая при слишком высоком уровне может привести к неприятному вкусу уксуса|
|citric acid|found in small quantities, citric acid can add 'freshness' and flavor to wines|Лимонная кислота, содержащаяся в небольших количествах, может придать винам «свежесть» и аромат.|
|residual sugar|the amount of sugar remaining after fermentation stops, it's rare to find wines with less than 1 gram/liter and wines with greater than 45 grams/liter are considered sweet|количество сахара, оставшееся после прекращения брожения, редко можно найти вина с содержанием сахара менее 1 грамма на литр, а вина с содержанием сахара более 45 граммов на литр считаются сладкими|
|chlorides|the amount of salt in the wine|количество соли в вине|
|free sulfur dioxide|the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion; it prevents microbial growth and the oxidation of wine|свободная форма SO2 существует в равновесии между молекулярным SO2 (в виде растворенного газа) и бисульфит-ионом; предотвращает рост микробов и окисление вина|
|total sulfur dioxide|amount of free and bound forms of S02; in low concentrations, SO2 is mostly undetectable in wine, but at free SO2 concentrations over 50 ppm, SO2 becomes evident in the nose and taste of wine|количество свободных и связанных форм S02; при низких концентрациях SO2 практически не обнаруживается в вине, но при концентрациях свободного SO2 более 50 частей на миллион SO2 становится заметным в аромате и вкусе вина|
|density|the density of water is close to that of water depending on the percent alcohol and sugar content|плотность воды близка к плотности воды в зависимости от процентного содержания спирта и сахара|
|pH|describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic); most wines are between 3-4 on the pH scale|описывает, насколько кислым или щелочным является вино по шкале от 0 (очень кислое) до 14 (очень щелочное); большинство вин имеют показатель рН от 3 до 4 по шкале pH.|
|sulphates|a wine additive which can contribute to sulfur dioxide gas (S02) levels, wich acts as an antimicrobial and antioxidant|добавка к вину, которая может способствовать повышению уровня сернистого газа (SO2), действующая как противомикробное и антиоксидантное средство.|
|alchohol|the percent alcohol content of the wine|процент содержания алкоголя в вине|
|quality|output variable (based on sensory data, score between 0 and 10)|выходная переменная (на основе сенсорных данных, оценка от 0 до 10)|
