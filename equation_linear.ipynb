```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Linear equation

```python
def calc_y(x, k=1, b=0):
    return k*x+b
```

```python
x = np.arange(-10, 11)
```

```python
plt.plot(x, calc_y(x, k=2, b=3), label='y = 2 * x + 3', color='r')
plt.plot(x, calc_y(x, k=3, b=3), label='y = 3 * x + 3', color='g')
plt.plot(x, calc_y(x, k=4, b=3), label='y = 4 * x + 3', color='b')
plt.grid()
plt.axvline(0, color='black', linestyle='--')
plt.axhline(0, color='black', linestyle='--')
plt.legend()
```

```python
plt.plot(x, calc_y(x, k=2, b=3), label='y = 2 * x + 3', color='r')
plt.plot(x, calc_y(x, k=2, b=8), label='y = 2 * x + 8', color='g')
plt.plot(x, calc_y(x, k=2, b=13), label='y = 2 * x + 13', color='b')
plt.grid()
plt.axvline(0, color='black', linestyle='--')
plt.axhline(0, color='black', linestyle='--')
plt.legend()
```
