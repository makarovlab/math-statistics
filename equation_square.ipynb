```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Square equation graph

```python
x = np.arange(-10, 11)
```

```python
plt.plot(x, x**2, color='r', label='y = x^2')
plt.plot(x, x**2 + 20, color='g', label='y = x^2 + 20')
plt.grid()
plt.axvline(0, color='black', linestyle='--')
plt.axhline(0, color='black', linestyle='--')
plt.legend(loc='upper left')
```

##### Derivative at each point

$\frac{dy}{dx} = \frac{y_2 - y_1}{x_2 - x_1}$

#### Example

$x_1, y_1 = (5, 25)$

$x_2, y_2 = (7.5, 56.25)$

$\frac{56.25 - 25}{7.5 - 5} = 12.5$

#### Example with small values
It shows why we have table of derivatives and limits:

$x_1, y_1 = (5, 25)$

$x_2, y_2 = (5.000001, 25.000010000001)$

$\frac{25.000010000001 - 25}{5.000001 - 5} = 10.000001 \approx 10 \approx 2 * 5$

```python
plt.plot(x, x**2, color='r', label='y = x^2')
plt.grid()
plt.axvline(0, color='black', linestyle='--')
plt.axhline(0, color='black', linestyle='--')
plt.legend(loc='upper left')

for n in x:
    plt.plot(n, n**2,  marker='o', markersize=6, color='b')
    plt.annotate(2*n, (n, n**2), xytext=(8, 0), textcoords='offset points')
```

Derivative sign shows in which direction function increases or decreases.
To minimize error we can take derivative with the opposite sign.

learning_rate = 0.001
derivative = -10
weight = 2.34

weight = weight - (derivative * learning_rate)
weight = 2.34 - (-10 * 0.001) = 2.34 + 0.01 = 2.35
