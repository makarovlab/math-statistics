#### Euclidean distance

Euclidean distance is commonly used in various applications, such as in clustering algorithms to group similar data points together, in machine learning to measure the similarity between two instances, and in image processing to calculate the distance between two pixels.

$D = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2 + ... + (n_2 - n_1)^2}$

```python
import matplotlib.pyplot as plt
import math
```

```python
def get_euclidean_distance(start=(3, 1), end=(4,5)):
    x_1, y_1 = start
    x_2, y_2 = end

    distance = math.sqrt((x_2 - x_1)**2 + (y_2 - y_1)**2)
    return round(distance, 3)
```

```python
def calculate_centroid(x: list[float], y: list[float]):
    c_x = sum(x) / len(x)
    c_y = sum(y) / len(y)

    return (c_x, c_y)
```

```python
def plot_distance(start=(3, 1), end=(1, 2)):

    A_x, A_y = start
    B_x, B_y = end

    # Calc center of line
    AB_mid_x = (A_x + B_x) / 2
    AB_mid_y = (A_y + B_y) / 2

    # Calc euclidean distance
    AB_distance = get_euclidean_distance(
        start=(A_x, A_y),
        end=(B_x, B_y)
        )

    plt.annotate(
        f"{AB_distance}",
        (AB_mid_x, AB_mid_y),
        xytext=(8, 0),
        textcoords='offset points'
        )
    
    plt.scatter(AB_mid_x, AB_mid_y, c='b')
    plt.plot([A_x, B_x], [A_y, B_y], 'b--')
```

```python
points = {
    "A": (3, 1),
    "B": (6, 4),
    "C": (4, 6),
    "D": (8, 7),
    "K": (4.5, 4.5),
}

distance_calc_task = [
    ["K", "B"],
    ["K", "C"],
    ["K", "D"],
    ["K", "A"],
]
```

```python
# Visualize points
for point in points.items():
    point_name, coords = point
    x_p, y_p = coords

    plt.scatter(x_p, y_p, c="red")
    plt.annotate(f"{point_name}({x_p}, {y_p})", (x_p, y_p), xytext=(4, 4), textcoords='offset points')

# Plot distance between A and B
for task in distance_calc_task:
    start, end = task
    plot_distance(start=points[start], end=points[end])
    

# Plot Centroid
x, y = list(zip(*[points["A"], points["B"], points["C"], points["D"]]))
x_c, y_c = calculate_centroid(x, y)
plt.scatter(x_c, y_c, c="g", s=65)

# Final plots
plt.grid()
```
