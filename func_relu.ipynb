#### ReLU (Rectified Linear Unit)

ReLU is an activation function commonly used in neural networks for machine learning. It is a simple, yet powerful function that is used to introduce non-linearity to the neural network.

The ReLU function is defined as follows: $f(x) = max(0, x)$

In other words, the function returns the input value if it is positive, and 0 otherwise. This means that the function is linear for positive inputs, and zero for negative inputs.

The ReLU function is widely used in deep learning because it has several desirable properties. First, it is computationally efficient to evaluate, making it well-suited for large-scale neural networks. Second, it has a simple derivative, which allows for efficient gradient-based optimization algorithms. Finally, the non-linear nature of the function allows neural networks to learn more complex and expressive representations of data.

```python
import matplotlib.pyplot as plt
```

```python
def relu(x):
    return x if x > 0 else 0
```

```python
relu(-13)
```

```python
X = list(range(-30, 21))
y = list(map(relu, X))
```

```python
plt.plot(X, y, label='Rectified Linear Unit')
plt.grid()
plt.axvline(0, c='black', linestyle='--')
plt.legend()
plt.title("Neuron activation function: ReLU")
```
