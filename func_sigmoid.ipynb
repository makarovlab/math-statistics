```python
import matplotlib.pyplot as plt
import math
```

```python
def sig(x):
    return 1 / (1 + math.e**-x)
```

```python
X = list(range(-10, 11))
y = list(map(sig, X))
```

```python
plt.plot(X, y, label='Sigmoid')
plt.axvline(0, c='black', linestyle='--')
plt.title("Neuron activation function: Sigmoid")
plt.legend()
plt.grid()
```
