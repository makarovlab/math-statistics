```python
import matplotlib.pyplot as plt
import math
```

```python
def sig(x):
    return 1 / (1 + math.e**-x)
```

```python
X = list(range(-4, 5))
y = list(map(sig, X))
```

```python
print("X", X)
print("y", y)
```

```python
plt.plot(X, y, label='Sigmoid')
plt.axvline(0, c='black', linestyle='--')
plt.title("Neuron activation function: Sigmoid")
plt.legend()
plt.grid()
```

```python
y_e = [1/(3 + math.e**-i) for i in X]
plt.plot(X, y_e)
```
