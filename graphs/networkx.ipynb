#### Graph is a data structure which helps to interpret complex relations between real world or abstract object

It helps to study, research structure and dynamics of social, biological and infrastructure networks.
Core elements of graph are nodes and edges. Node represent objects like:
- user of facebook
- start or end of road
- neuron cell in a human brain
- website
- html tag inside inside webpage
- city, country, address, place
- stars, planets
- words, sentences, articles
- goods in the store
- ...

Nodes can have connections to other nodes through edges. Or in other words relations:
- user of facebook has friends and his friends also have friends;
- start of a road connects to end and end of a road connects to another start of road;
- neuron cell connects to another neuron trough axon;
- website contains links to other websites;
- html tag inside webpage contains child tags which also contain child tags;
- country contains regions, regions contain cities, cities contain addresses
- starts have planets
- articles contain sentences, sentences contain words, words contain letters
- stores contain goods

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_edges_from([(1, 2), (1,3), (2, 3), (3,5), (2, 4)])

nx.draw(G, with_labels=True)
```
