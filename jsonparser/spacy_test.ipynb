```python
import spacy
import pandas as pd
import numpy as np
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df = pd.read_csv("words.csv")
```

```python
df
```

```python
labels = ['ref', 'address', 'housenumber', 'street', 'city', 'state', 'country', 'postcode', 'email', 'opening_hours', 'phone', 'lat', 'lon', 'location']
lbs = [nlp(label) for label in labels]
```

```python
for word in df['text']:
    doc = nlp(word)
    index = np.argmax([doc.similarity(lb) for lb in lbs])
    print(f"Word: {word}; Matched: {labels[index]}")
```
