```python
import tiktoken
import pandas as pd
```

```python
df = pd.read_csv("words2.csv")
df.head()
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
tokens = encoding.encode("tiktoken is great!")
print(tokens)


[encoding.decode_single_token_bytes(token) for token in tokens]
```

```python
def tokenize(input_string):
    tokens = encoding.encode(input_string)
    return [encoding.decode_single_token_bytes(token).decode("utf-8") for token in tokens]
```

```python
for index, row in df.iterrows():
    cleaned_text = row['text'].lower().replace("_", " ")
    print(tokenize(cleaned_text))
```
