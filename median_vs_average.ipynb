```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
salaries = np.array([
    1800,
    2000,
    2030,
    2320,
    3400,
    4440,
    5550,
    5600,
    6600,
    8600,
    20000,
    25000
])
```

```python
s_average = np.mean(salaries)
s_median = np.median(salaries)

print(f"Average: {s_average}; Median: {s_median}")
```

```python
plt.hist(salaries)
plt.axvline(s_average, color='r', linestyle='-')
plt.axvline(s_median, color='orange', linestyle='-')
```

```python
no_extreme_vals = salaries[0:8]
plt.hist(no_extreme_vals)
plt.axvline(np.mean(no_extreme_vals), color='r', linestyle='-')
plt.axvline(np.median(no_extreme_vals), color='orange', linestyle='-')
```
