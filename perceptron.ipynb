#### Perceptron

Basib math formula:
$f(x) = \sum_i^nw_ix_i$

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
def neuron(X: list[float], W: list[float]):
    return np.dot(X, W)
```

```python
X = np.random.randint(0, 10, size=(100, 2))
Y = np.array([a + b for a, b in X])
```

```python
def optimizer(y, y_pred, weights):
    alpha = 0.001
    delta = y - y_pred

    if delta > 0:
        weights += delta * weights * alpha
    else:
        weights -= delta * weights * alpha
    
    return weights
```

```python
weights = np.random.rand(2)
```

```python
for epoch in range(0, 2):
    print(f"Epoch #{epoch};")

    for i, value in enumerate(X):
        prediction = neuron(X[i], weights)
        new_weights = optimizer(prediction, Y[i], weights)
        weights = new_weights

        print(f"{X[i][0]} + {X[i][1]} = {Y[i]}; Prediction: {prediction}; Delta: {prediction - Y[i]}")
```

```python
print(f"MSE: {mean_squared_error(Y, [neuron(x, weights) for x in X])}")
print(f"MSE: {mean_absolute_error(Y, [neuron(x, weights) for x in X])}")
print(f"MSE: {r2_score(Y, [neuron(x, weights) for x in X])}")
```
