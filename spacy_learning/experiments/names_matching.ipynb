```python
import spacy
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://ru.wikipedia.org/wiki/%D0%93%D0%BE%D1%81%D1%83%D0%B4%D0%B0%D1%80%D1%81%D1%82%D0%B2%D0%B5%D0%BD%D0%BD%D0%B0%D1%8F_%D0%B4%D1%83%D0%BC%D0%B0")
```

```python
soup = BeautifulSoup(response.text)
corpus = soup.text.replace("\n", "")
```

```python
nlp = spacy.load("ru_core_news_md")
```

```python
doc = nlp(corpus)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
